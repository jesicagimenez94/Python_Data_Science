```python
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d.axes3d import Axes3D, get_test_data
from matplotlib import cm
import numpy as np

data= pd.read_csv('C:/Users/jesica/Downloads/Forbes.csv') #agregar la base de datos

x=data['Sales']
y=data['Market Value']
z=data['Profits']

zy=[2,2]

fig = plt.figure(figsize=plt.figaspect(0.2))
grafica=fig.add_subplot(1,2,1, projection='3d')
grafica.scatter3D(x,y,z,c=z,cmap='Dark2')
grafica.set_xlabel('Sales')
grafica.set_ylabel('Market Value')
grafica.set_zlabel('Profits')
plt.title('Grafico de Dispersion') 



grafica = fig.add_subplot(1, 2, 2, projection='3d')
x, y, zy = get_test_data(0.09)
grafica.plot_wireframe(x, y, zy, rstride=10, cstride=10,label='Curve')
grafica.set_xlabel('Sales')
grafica.set_ylabel('Market Value')
grafica.set_zlabel('Profits')
plt.title('Grafico Subtramado')
grafica.legend()

plt.show()
```
